# Regresión lineal múltiple

En este archivo se desarrolla el algoritmo de **regresión lineal múltiple** en el cual se emplean las características 'Seq', 'M', 'R', 'Teff','L','Meta','logg' y 'Prot' para estimar la edad estelar.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/jarmireciomartinez/Doctorado_Python/Pruebas_regresion/gyro_tot_v20180801.txt', sep="\t", header=0)
df
```

```python
# Se eliminan las filas con valores NULL/NA

df = df.dropna()
```

```python
df_2 = df[['Seq', 'M', 'R', 'Teff','L','Meta','logg','Prot','Age']]
df_2
```

```python
#Definición de las variables a emplear

Y_mult = np.array(df_2['Age']) #Predecir Age
X_mult = np.array(df[['Seq', 'M', 'R', 'Teff','L','Meta','logg','Prot']]) #A través de las demás columnas
```

```python
#Se separa el 20% de los datos para test

X_train, X_test, Y_train, Y_test = train_test_split(X_mult, Y_mult, test_size=0.2)
```

```python
#Definición del algoritmo
regr_mult = linear_model.LinearRegression()

#Entrenamiento del algoritmo
regr_mult.fit(X_train,Y_train)
```

```python
#Realizar una predicción
Y_pred = regr_mult.predict(X_test)
```

```python
#Comparación de arrays

data = {'Y_test':  Y_test,
        'Y_pred': Y_pred,
        }

df_new = pd.DataFrame (data, columns = ['Y_test','Y_pred'])

print (df_new)
```

```python
#Valores de pendiente e intersección

print('Valor de la pendiente:')
print(regr_mult.coef_)
print('Valor de la intersección:')
print(regr_mult.intercept_)
```

```python
#Precisión del algoritmo (conjunto entrenamiento)
print('Precisión del algoritmo:')
print(regr_mult.score(X_train, Y_train))
```

```python
#Precisión del algoritmo (conjunto test)
print('Precisión del algoritmo:')
print(regr_mult.score(X_test, Y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score, max_error

# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(Y_test, Y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(Y_test, Y_pred))
# The max error
print('Max error: %.2f'
      % max_error(Y_test, Y_pred))
```
