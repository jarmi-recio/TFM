# Descenso de Gradiente Estocástico

En este archivo se desarrolla el algoritmo de **SGD** en el cual se emplean las características 'Seq', 'M', 'R', 'Teff','L','Meta','logg' y 'Prot' para estimar la edad estelar.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/jarmireciomartinez/Doctorado_Python/Pruebas_regresion/gyro_tot_v20180801.txt', sep="\t", header=0)
df
```

```python
# Se eliminan las filas con valores NULL/NA

df = df.dropna()
```

```python
df_2 = df[['Seq', 'M', 'R', 'Teff','L','Meta','logg','Prot','Age']]
df_2
```

```python
#Definición de las variables a emplear

Y_sgd = np.array(df_2['Age']) #Predecir Age
#X_sgd = np.array(df[['Seq', 'M', 'R', 'Teff','L','Meta','logg','Prot']]) #A través de las demás columnas
X_sgd = np.array(df_2[['M']]) #A través de la masa

from sklearn import preprocessing
from sklearn.preprocessing import normalize

X_norm = preprocessing.normalize(X_sgd, norm='l2') #Normalización del conjunto de características
```

```python
#Se separa el 20% de los datos para test

X_train, X_test, Y_train, Y_test = train_test_split(X_sgd, Y_sgd, test_size=0.2)
```

```python
#Definición del algoritmo
sgd_regr = SGDRegressor()

#Entrenamiento del algoritmo
sgd_regr.fit(X_train,Y_train)
```

```python
#Realizar una predicción

Y_pred = sgd_regr.predict(X_test)
```

```python
#Comparación de arrays

data = {'Y_test':  Y_test,
        'Y_pred': Y_pred,
        }

df_new = pd.DataFrame (data, columns = ['Y_test','Y_pred'])

print (df_new)
```

```python
#Precisión del algoritmo (conjunto entrenamiento)

print('Precisión del algoritmo:')
print(sgd_regr.score(X_train, Y_train))
```

```python
#Precisión del algoritmo (conjunto test)

print('Precisión del algoritmo:')
print(sgd_regr.score(X_test, Y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score, max_error, mean_absolute_error

# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(Y_test, Y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(Y_test, Y_pred))
# The max error
print('Max error: %.2f'
      % max_error(Y_test, Y_pred))
# The mean absolute error
print('MAE: %.2f'
      % mean_absolute_error(Y_test, Y_pred))
```
